```python
# Challenge 1: Square Numbers and Return Their Sum.

class Point:

    def __init__(self,x,y,z):
        self.x = x
        self.y = y
        self.z = z

    def sqSum(self):
      return self.x**2+self.y**2+self.z**2

a = Point(1,3,5)
a.sqSum() 
```

```python
# Challenge 5: Handling a Bank Account.


class Account:
    def __init__(self, title=None, balance=0):
        self.title = title
        self.balance = balance
    
    def deposit(self, amount):
        self.balance = self.balance+amount
        print("Balance deposit is:",self.balance)

    def withdraw(self, amount):
        self.balance = self.balance-amount
        print("Balance after withdraw is:",self.balance)


    def getBalance(self):
        return self.balance

class SavingsAccount(Account):
    def __init__(self, title=None, balance=0, interestRate=0):
            super().__init__(title, balance)
            self.interestRate = interestRate
    
    def interestAmount(self):
        self.interestAmount = (self.interestRate*self.balance)/100
        return int(self.interestAmount)

s = SavingsAccount("Ashish",2000,5)
s.deposit(500)
s.withdraw(500)
print("Total Balance Left :",s.getBalance())
print("Interest Amount is :",s.interestAmount())
```
